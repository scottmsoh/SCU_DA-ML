### Q3.4, Based on top 5 trading volume dates for stock closing price over 100 dollars, which date it gains the most money per share? Assuming buy at open and sell at close.(choose the date in Camino)

```python
above_100_data = df[df['Close'] > 100]
above_100_data
```

```python
top_5_volumes3 = above_100_data.nlargest(5, 'Volume')
top_5_volumes3
```

```python
top_5_volumes3['Gain'] = top_5_volumes3['Close'] - top_5_volumes3['Open']
top_5_volumes3['Gain']
```

```python
max_gain_date = top_5_volumes3[top_5_volumes3['Gain'] == top_5_volumes3['Gain'].max()]['Date'].iloc[0]
max_gain_date
```
